# Merging, Joining & Concatenation

```python



# Netflix dataset (df1)
df1 = pd.DataFrame({
    "show_id": ["s1", "s2", "s3", "s4"],
    "title": ["Inception", "Titanic", "Interstellar", "Avatar"],
    "release_year": [2010, 1997, 2014, 2009],
    "country": ["USA", "USA", "SOUTH KOREA", "INDIA"],
    "rating": ["PG-13", "PG-13", "PG-13", "PG-13"]
})

# Ratings dataset (df2)
df2 = pd.DataFrame({
    "title": ["Inception", "Titanic", "Interstellar", "The Matrix"],
    "average_rating": [8.8, 7.8, 8.6, 8.7],
    "country": ["USA", "USA", "SOUTH KOREA", "INDIA"],
    "total_reviews": [1200000, 900000, 1100000, 1500000]
})
```

***40. Merge datasets***

```python
df_merged = df1.merge(df2, on="title", how="inner")

print(df_merged)
```

***41. Left join***

```python
df_merged=df1.merge(df2, on= "title", how="left")
print(df_merged)
```

***42. Append rows***

```python
df_combined = pd.concat([df1, df2], ignore_index=True)
print(df_combined)
```
